### Repeat the following steps for each column within outliers

* Visualize the outliers
* Detect if your data set contains any outliers
* Identify lower & higher limit of outliers
* Drop all outliers
* Visualize the column after processing


```python
sns.pairplot(df[['PRCP','TMAX','TMIN']]);
```

```python
sns.boxplot(df[['PRCP','TMAX','TMIN']])
```

```python
col = df.select_dtypes(include=[np.number])
Q1 = col.quantile(0.25)
Q3 = col.quantile(0.75)
IQR = Q3 - Q1

outlier = 1.5

lower = Q1 - outlier * IQR
upper = Q3 + outlier * IQR

df_outlier = pd.DataFrame({'Lower Limit': lower, 'Upper Limit': upper})

print("Lower and Upper Limits for Outliers:\n")
print(df_outlier)
```

```python
clean = df[(col >= lower) & (col <= upper)]

print("DataFrame without outliers:")
print(clean.head(5))
```

```python
sns.pairplot(clean[['PRCP','TMAX','TMIN']]);
```
